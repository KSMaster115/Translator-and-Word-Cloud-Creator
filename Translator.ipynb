```python
df = pd.read_excel(r"Responses.xlsx", sheet_name = "A1")
```

```python
print(df.columns.get_loc('Q6_a_Pink (UA)'))
```

```python
df.drop(df.iloc[:,5:65], axis = 1, inplace = True)
df.drop(df.iloc[:,2:4], axis = 1, inplace = True)
```

```python
APAC = df.hCountry <= 3
EUROPE = ((4 <= df.hCountry) & (df.hCountry <= 6)) | (df.hCountry == 10)
LATAM = (7 == df.hCountry) | (df.hCountry == 8)
AMERICA = df.hCountry == 9
condlist = [APAC, EUROPE, LATAM, AMERICA]
choicelist = ['APAC', 'EUROPE', 'LATAM', 'AMERICA']
df['Regions'] = np.select(condlist, choicelist)
movingcol = df.pop('Regions')
df.insert(3, "Regions", movingcol)
```

```python
df.fillna(0, inplace = True)
```

```python
df
```

```python
for r in range(4,36):
    name = df.columns[r]
    temp = [ts.translate_text(i, translator='google', sleep_seconds=0.5) if isinstance(i, str) else i for i in df.iloc[:,r]]
    df.drop(name, axis = 1, inplace = True)
    df.insert(r, name, temp)
```
